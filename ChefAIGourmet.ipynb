```python
import google.generativeai as genai
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display, clear_output

# Configura√ß√£o API_KEY
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

generation_config = {
    "candidate_count": 1,
    "temperature": 0.8,
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}

model = genai.GenerativeModel(
    model_name="gemini-1.5-pro-latest",
    generation_config=generation_config,
    safety_settings=safety_settings
)

def gerar_receita(button):
    generate_button.disabled = True
    ingredientes = text_input.value
    if ingredientes:
        prompt = f"Crie uma receita simples com os seguintes ingredientes: {ingredientes}"
        resposta = model.generate_content(prompt)
        output_area.clear_output()
        if resposta.text.strip():  # Verificando conte√∫do
            with output_area:
                print("üìñ Receita Sugerida:")
                print("------------------------------")
                print(resposta.text)
                print("------------------------------")
        else:
            with output_area:
                print("‚ùå N√£o foi poss√≠vel encontrar uma receita para os ingredientes fornecidos.")
    else:
        with output_area:
            clear_output()
            print("üö´ Por favor, insira ao menos um ingrediente.")
    generate_button.disabled = False

# Cabe√ßalho e mensagem de boas-vindas
title = widgets.HTML("<h1>üç≥ Gerador de Receitas üìù</h1>")
welcome_message = widgets.HTML("<h3>üëã Bem-vindo ao gerador de receitas! Insira os ingredientes para come√ßar.</h3>")

# Descri√ß√£o dos ingredientes em linha separada
ingredients_label = widgets.Label('ü•ï Ingredientes:')

# Interface de entrada de dados
text_input = widgets.Text(
    placeholder='Digite os ingredientes separados por v√≠rgula',
    layout={'width': '600px'}
)

generate_button = widgets.Button(
    description='Gerar Receita',
    button_style='info',
    icon='check'
)

output_area = widgets.Output()

generate_button.on_click(gerar_receita)

# Espa√ßos entre os elementos
space_between_text_and_button = widgets.HTML("<br>")
space_between_button_and_output = widgets.HTML("<br>")

# Montando a interface
box_layout = widgets.VBox([
    title,
    welcome_message,
    ingredients_label,
    text_input,
    space_between_text_and_button,
    generate_button,
    space_between_button_and_output,
    output_area
])
display(box_layout)
```
